```python
import torch
import torch.nn as nn
# 定义一维卷积层
# conv1d = nn.Conv1d(in_channels=4, out_channels=1, kernel_size=3)
conv1d = nn.Conv1d(in_channels=3, out_channels=3, kernel_size=3, stride=2, padding=1) 
conv1d.weight.data = torch.ones_like(conv1d.weight.data)
conv1d.bias.data = torch.zeros_like(conv1d.bias.data)
# 创建输入数据
input_data = torch.arange(75).reshape(1, 25, 3).float() # batch_size x text_len x embedding_size
# 调整输入数据的维度以匹配卷积层的期望格式
input = input_data.permute(0, 2, 1) # batch_size x embedding_size x text_len
# 应用一维卷积层
output = conv1d(input)
# 输出结果的维度
print(output.size()) # 输出: torch.Size([32, 100, 34])

```

```python
convTrans1d = nn.ConvTranspose1d(3, 3, kernel_size=4, stride=2, padding=1)
outputTrans = convTrans1d(output)
outputTrans.size()
```
